```python
import matplotlib.pyplot as plt

def LossReader(file_path):
    with open(file_path, "r") as file:
        lines = file.readlines()

    log_file = []
    for line in lines:
        log_file.append(line.split())

    recon_losses = [x for x in log_file if 'count=20000' in x]

    list_total_loss = []

    for recon_loss in recon_losses:
        print(recon_loss)
        list_total_loss.append(float(recon_loss[2]))
    print(len(list_total_loss))
    plt.plot(list_total_loss)
    plt.xlabel('Index')
    plt.ylabel("total_loss")
    plt.title('Rec Loss Plot')
    plt.show()

    return None
```

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# 간단한 모델 정의
class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.conv1 = nn.Conv2d(1, 10, kernel_size=5)
        self.conv2 = nn.Conv2d(10, 20, kernel_size=5)
        self.fc1 = nn.Linear(320, 50)
        self.fc2 = nn.Linear(50, 10)

    def forward(self, x):
        x = torch.relu(torch.max_pool2d(self.conv1(x), 2))
        x = torch.relu(torch.max_pool2d(self.conv2(x), 2))
        x = x.view(-1, 320)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# 모델 인스턴스 생성
model = resnet18(pretrained=True)

# 각 레이어별 파라미터 수 수집
layer_names = []
param_counts = []

for name, param in model.named_parameters():
    if param.requires_grad:
        layer_names.append(name)
        param_counts.append(param.numel())

# 파라미터 수를 막대 차트로 시각화
plt.figure(figsize=(10, 15))
plt.barh(layer_names, param_counts, color='skyblue')
plt.xlabel('Number of Parameters')
plt.ylabel('Layer Name')
plt.title('Number of Parameters per Layer')
plt.gca().invert_yaxis()  # 레이어 순서 위에서 아래로
plt.show()
```

```python
import torch
import torch.nn as nn


# 간단한 모델 정의
class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.conv1 = nn.Conv2d(1, 10, kernel_size=5)
        self.conv2 = nn.Conv2d(10, 20, kernel_size=5)
        self.fc1 = nn.Linear(320, 50)
        self.fc2 = nn.Linear(50, 10)
        self.submodule = nn.Sequential(nn.Conv2d(20, 40, kernel_size=3), nn.ReLU())

    def forward(self, x):
        x = torch.relu(torch.max_pool2d(self.conv1(x), 2))
        x = torch.relu(torch.max_pool2d(self.conv2(x), 2))
        x = x.view(-1, 320)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x


# 모델 인스턴스 생성
model = SimpleModel()

# 각 파라미터 및 서브모듈의 부모 모듈 이름 출력
for name, param in model.named_parameters():
    parent_module_name = ".".join(name.split(".")[:-1])  # 마지막 부분을 제외한 나머지
    print(f"Parameter: {name}, Parent module: {parent_module_name}")

for name, module in model.named_modules():
    if name:  # 이름이 빈 문자열이 아닌 경우에만
        parent_module_name = ".".join(
            name.split(".")[:-1]
        )  # 마지막 부분을 제외한 나머지
        print(f"Module: {name}, Parent module: {parent_module_name}")
```
